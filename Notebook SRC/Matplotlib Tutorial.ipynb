```python
batsman = pd.read_csv("./sharma-kohli.csv")
```

```python
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange')
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green', linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange', linestyle='dashed')
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green', linestyle='dotted')
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange', linestyle='dotted')
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green', linewidth=4)
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange', linewidth=4)
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green', linewidth=2, marker='o')
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange', linewidth=2, marker='o')
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='Green', linewidth=2, marker='o', label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='orange', linewidth=2, marker='o', label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.legend(loc='upper left')
```
