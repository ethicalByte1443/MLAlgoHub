# ***Option 2. Fill missing values with Scikit Learn***

```python
car_sales_missing = pd.read_csv("./car-sales-extended-missing-data.csv") 
```

```python
car_sales_missing.isna().sum()
```

```python
car_sales_missing.dropna(subset=['Price'], inplace=True)
```

```python
car_sales_missing.isna().sum()
```

```python
x = car_sales_missing.drop('Price', axis=1)
x
```

```python
y = car_sales_missing['Price']
```

```python
y
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.compose import ColumnTransformer
```

```python
# Fill categorical value with --> 'missing'
# Fill numerical vlaues with --> mean()
```

```python
cat_imputer = SimpleImputer(strategy='constant', fill_value='missing')
door_imputer = SimpleImputer(strategy='constant', fill_value=4)
numerical_imputer = SimpleImputer(strategy='mean')
```

```python
# Define Columns
cat_features = ['Make', 'Colour']
door_features = ['Doors']
num_features = ['Odometer (KM)']
```

```python
# Create Imputer (Something that fills missing data)
imputer = ColumnTransformer([
    ('cat_imputer', cat_imputer, cat_features),
    ('door_imputer', door_imputer, door_features),
    ('numerical_imputer', numerical_imputer, num_features)
])

#  tronsform the data
filled_x = imputer.fit_transform(x)
filled_x
```

```python
car_sales_missing = pd.DataFrame(filled_x, columns=["Make", "Colour", "Doors", "Odometer (KM)"])
```

```python
car_sales_missing
```

```python
car_sales_missing.isna().sum()
```

```python
x = car_sales_missing
x
```

```python
len(y)
```

```python
categorical_features = ['Make', 'Colour', 'Doors']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot',
                                   one_hot,
                                   categorical_features)],
                                 remainder='passthrough')
transform_x = transformer.fit_transform(x)
transform_x
```

```python
# NOw we've got our data as numbers and filled (no missing)
np.random.seed(42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection  import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(transform_x, y, test_size=0.2)
```

```python
clf = RandomForestRegressor(n_estimators=100)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
clf.score(x_train, y_train)
```
