```python
import pandas as pd
```

```python
# Read the original CSV file where rows are separated by semicolons
file_path = 'cardio_train.csv'  # Update with your file path
df = pd.read_csv(file_path, header=None, delimiter=';')

# Define the column names
columns = ['id', 'age', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']

# Assign column names to the dataframe
df.columns = columns

# Save the new CSV with separate columns
new_file_path = 'formated_cardio_train.csv'  # The new file with separated columns
df.to_csv(new_file_path, index=False)

print(f"New CSV file created: {new_file_path}")
```

```python
data = pd.read_csv("formated_cardio_train.csv")
```

```python
data = data.drop('id', axis=1)
data = data.drop('age', axis = 1)
```

```python
y = data['cardio']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100)
```

```python

print(data.dtypes)
```

```python

# Identify non-integer values in each column
for col in data.columns:
    invalid_values = data[col].str.isnumeric() == False
    if invalid_values.any():
        print(f"Column: {col}")
        print(data.loc[invalid_values, col])
```

```python
data = data.apply(lambda x: pd.to_numeric(x.str.extract('(\d+)')[0], errors='coerce'))
```

```python
data = data.apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
```

```python
numeric_cols = ['gender','height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']
data[numeric_cols] = data[numeric_cols].astype(int)
```

```python
print(data.dtypes)  # Ensure all required columns are integers
print(data.head())  # Check the cleaned DataFrame
```

```python
for i in range(10, 200, 10):
    print(f"n_estimators = {i} :")  # Using f-string to format the string with the value of i
    clf = RandomForestClassifier(n_estimators=i)
    clf.fit(x_train, y_train)
    score = clf.score(x_test, y_test)
    print(f"{score * 100 : .2f}")
    
```
