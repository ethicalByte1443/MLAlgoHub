```python
import pandas as pd
import numpy as np
```

```python
car_sales = pd.read_csv('./car-sales-extended-missing-data.csv')
```

```python
car_sales
```

```python
car_sales.isna().sum()
```

```python
car_sales['Make'].fillna('missing', inplace = True)
```

```python
car_sales['Colour'].fillna('missing', inplace=True)
```

```python
car_sales['Odometer (KM)'].fillna(car_sales['Odometer (KM)'].mean(), inplace=True)
```

```python
car_sales['Doors'].value_counts()
```

```python
car_sales['Doors'].fillna(4, inplace=True)
```

```python
car_sales.dtypes
```

```python
car_sales.isna().sum()
```

```python
car_sales.dropna(inplace=True)
```

```python
car_sales
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
x = car_sales.drop('Price', axis=1)
```

```python
x['Odometer (KM)'] = x['Odometer (KM)'].astype('int64')
```

```python
x['Doors'] = x['Doors'].astype('int64')
```

```python
x
```

```python
car_sales
```

```python
y = car_sales['Price'].astype('int64')
```

```python


# Ensure the input is a DataFrame for consistent output
categorical_features = ['Make', 'Colour', 'Doors']

one_hot = OneHotEncoder(sparse_output=False)  # Force dense output
transformer = ColumnTransformer([('one_hot',
                                 one_hot,
                                 categorical_features)],
                               remainder='passthrough')

transformed_x = transformer.fit_transform(x)  # x should be a DataFrame
print(transformed_x)
```

```python
pd.DataFrame(transformed_x)
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, train_size=0.8)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
